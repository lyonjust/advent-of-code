```python
import pandas as pd
import numpy as np
```

```python
df_input = pd.read_csv('input.txt', header=None)
```

```python
df_input['Plane'] = df_input[0].str.split(' ', 1)
```

```python
df_input['Direction'] = df_input['Plane'].str[0]
df_input['Magnitude'] = df_input['Plane'].str[1].astype(int)
```

```python
df_input['Sign'] = np.where(df_input['Direction'] == 'up', -1, 1)
df_input['Vector'] = df_input['Sign'] * df_input['Magnitude']
```

```python
df_vertical = df_input[df_input['Direction'] != 'forward']
df_horizontal = df_input[df_input['Direction'] == 'forward']
```

```python
total_vertical_displacement = df_vertical['Vector'].sum()
total_horizontal_displacement = df_horizontal['Vector'].sum()
```

```python
print(f'Final horizontal position {total_horizontal_displacement}, final depth {total_vertical_displacement}')
```

```python
answer_one = total_vertical_displacement * total_horizontal_displacement
answer_one
```

```python
df_input['AimChanged'] = np.where(df_input['Direction'] != 'forward', 1, 0)
df_input['Aim'] = df_input['AimChanged'] * df_input['Vector']
df_input['AimRunningTotal'] = df_input['Aim'].cumsum()
df_input
```

```python
df_input['Depth'] = np.where(df_input['Direction'] == 'forward', df_input['AimRunningTotal'] * df_input['Magnitude'], 0)
df_input
```

```python
new_depth = df_input['Depth'].sum()
answer_two = total_horizontal_displacement * new_depth
answer_two
```
